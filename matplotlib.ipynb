# bar plot

```python
import matplotlib.pyplot as plt 

x=["python","c","c++","java"]
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)
c=["y","b",'m',"g"]

plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt 

x=["python","c","c++","java"]
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)
c=["y","b",'m',"g"]

plt.bar(x,y,width=0.4,color="r",edgecolor="r",label="popularity")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]



plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)


plt.barh(x,y,color="r",label="popularity")
plt.barh(x,z,color="y",label="popularity1")


plt.xticks(rotation=20)
plt.legend()
plt.show()
```

# stack plot

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]


plt.stackplot(x,area,area1,area2)
plt.show()
```

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l)


plt.title("python")
plt.xlabel("x-axies")
plt.ylabel("y-axies")
plt.legend()
plt.show()
```

# fill_between plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y ,color="r")

plt.fill_between(x=[2,4],y1=2,y2=4,color="g")


plt.title("python")
plt.xlabel("x-axies")

plt.ylabel("y-axies")

```
