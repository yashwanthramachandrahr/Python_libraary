# Find the number of missing values in each column

```python
df.isnull().sum()
```

# Fill the missing values in the Rating column with the mean rating and show the updated DataFrame

```python
df["Rating"].fillna(df["Rating"].mean(),inplace=True)

df
```

# Display all products whose price is greater than the average price.

```python
avg_price=df["Price"].mean()

result=df[df["Price"]>avg_price]


result
```

# Find the average price for each product category.

```python
avg_price_category=df.groupby("Category")["Price"].mean()

avg_price_category
```

# Identify the category with the highest average price.

```python
avg_price_category=df.groupby("Category")["Price"].mean()

highest_avg_category=avg_price_category.idxmax()

highest_avg_category
```

# Count how many products are present in each category.

```python
category_count=df["Category"].value_counts()

category_count
```

# For each category, find the product with the highest price using idxmax().

```python
idx=df.groupby("Category")["Price"].idxmax()

highest_price_products=df.loc[idx]

highest_price_products
```

# For every Category–Brand combination, find the product with the highest price.

```python
idx=df.groupby(["Category","Brand"])["Price"].idxmax()

highest_price=df.loc[idx]

highest_price
```

```python
idx=df.groupby(["Category","Brand"])["Price"].idxmax()
result=df.loc[idx,["Category","Brand","Price"]]

result
```

# For each category, compute the following:

# Number of products , Average price , Total stock

```python
category_stats =df.groupby("Category").agg(
number_of_products=("Product ID","count"),
averages_price=("Price","mean"),
Total_stock=("Stock","sum")
)

category_stats
```

# For each brand, compute:

# Number of products

```python
brand_count=df["Brand"].value_counts()

brand_count
```

# Maximum price for each brand

```python
max_price=df.groupby("Brand")["Price"].max()
max_price
```

# Minimum stock for each brand

```python
min_stock=df.groupby("Brand")["Stock"].min()
min_stock
```

# Create a new column Total_Sales = Price * Stock.

```python
Total_sales=df["Price"] * df["Stock"]

Total_sales
```

# Group by brand and calculate the total sales for each brand

```python
# Step 1: Create Total_Sales column
df["Total_Sales"] = df["Price"] * df["Stock"]

# Step 2: Group by brand to calculate total sales
brand_total_sales = df.groupby("Brand")["Total_Sales"].sum()

print(brand_total_sales)
```

# Identify the brand with the highest total sales

```python
highest_sales_brand = brand_total_sales.idxmax()
highest_sales_brand
```

# Find how many unique brands are present within each category

```python

unique_brands_per_category = df.groupby("Category")["Brand"].unique()

unique_brands_per_category
```

# Display the top 3 categories based on the highest average price

```python
top3_categories = df.groupby("Category")["Price"].mean().sort_values(ascending=False).head(3)
print(top3_categories)
```
